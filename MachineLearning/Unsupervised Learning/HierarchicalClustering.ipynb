```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
data = pd.read_csv("Datasets/musteriler.csv")

x = data.loc[:,["Hacim", "Maas"]]
```

```python
ac = AgglomerativeClustering(n_clusters=3, metric="euclidean", linkage="ward")
tahmin = ac.fit_predict(x)
tahmin
```

```python
plt.scatter(x.iloc[tahmin==0,0], x.iloc[tahmin==0,1],s=100,c="red")
plt.scatter(x.iloc[tahmin==1,0], x.iloc[tahmin==1,1],s=100,c="blue")
plt.scatter(x.iloc[tahmin==2,0], x.iloc[tahmin==2,1],s=100,c="green")
```

```python
dendogram = sch.dendrogram(sch.linkage(x, method="ward"))
```
