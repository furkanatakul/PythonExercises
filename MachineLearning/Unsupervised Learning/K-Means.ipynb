```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("Datasets/musteriler.csv")

x = data.loc[:,["Hacim", "Maas"]]
```

```python
kmeans = KMeans(n_clusters=4, init="k-means++")
#kmeans.fit(x)
tahmin = kmeans.fit_predict(x)
kmeans.cluster_centers_
```

```python
plt.scatter(x.iloc[tahmin==0,0], x.iloc[tahmin==0,1],s=100,c="red")
plt.scatter(x.iloc[tahmin==1,0], x.iloc[tahmin==1,1],s=100,c="blue")
plt.scatter(x.iloc[tahmin==2,0], x.iloc[tahmin==2,1],s=100,c="green")
plt.scatter(x.iloc[tahmin==3,0], x.iloc[tahmin==3,1],s=100,c="black")
```

```python
results = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=123)
    kmeans.fit(x)
    results.append(kmeans.inertia_) #WCSS
plt.plot(range(1,11), results)
```
