```python
# veri yukleme
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
veriler = pd.read_csv('Datasets/maaslar.csv')

x = veriler.iloc[:,1:2]
y = veriler.iloc[:,2:]
X = x.values
Y = y.values
Z = X+0.5
K = X-0.5
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=0)
```

```python
dtr.fit(X,Y)
```

```python
plt.scatter(X,Y,color = "red")
plt.plot(X,dtr.predict(X), color = "blue")
plt.plot(X,dtr.predict(Z), color="green")
plt.plot(X,dtr.predict(K), color="orange")
```

```python
dtr.predict([[20]])
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=0, n_estimators=10)
rfr.fit(X,Y)
```

```python
rfr.predict([[6.6]])
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X,rfr.predict(X), color="blue")
plt.plot(X,rfr.predict(Z), color="green")
plt.plot(X,rfr.predict(K), color="orange")
```

```python
from sklearn.metrics import r2_score
print("Random Forest R2 değerleri:")
r2_score(Y,rfr.predict(X))
```

```python
r2_score(Y,rfr.predict(Z))
```

```python
r2_score(Y,rfr.predict(K))
```

```python
print("Decision Tree R2 değerleri:")
r2_score(Y,dtr.predict(X))
```

```python
r2_score(Y,dtr.predict(Z))
```

```python
r2_score(Y,dtr.predict(K))
```
